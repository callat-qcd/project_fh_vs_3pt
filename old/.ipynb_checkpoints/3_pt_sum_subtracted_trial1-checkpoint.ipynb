```python
# Get the data
file = h5py.File("a12m310_a_avg.h5", "r")

# 2-point
data_proton_dn_2pt = np.array(file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['spec']['ml0p0126']['proton']['px0_py0_pz0']['spin_dn'])
data_proton_up_2pt = np.array(file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['spec']['ml0p0126']['proton']['px0_py0_pz0']['spin_up'])
data_np_proton_dn_2pt = np.array(file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['spec']['ml0p0126']['proton_np']['px0_py0_pz0']['spin_dn'])
data_np_proton_up_2pt = np.array(file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['spec']['ml0p0126']['proton_np']['px0_py0_pz0']['spin_up'])


# 3-point
data_proton_DD_dn_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_DD_dn_dn_tsep_{}_sink_mom_px0_py0_pz0'.format(i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])
data_proton_DD_up_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_DD_up_up_tsep_{}_sink_mom_px0_py0_pz0'.format(i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])
data_np_proton_DD_dn_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_np_DD_dn_dn_tsep_{}_sink_mom_px0_py0_pz0'.format(-i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])
data_np_proton_DD_up_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_np_DD_up_up_tsep_{}_sink_mom_px0_py0_pz0'.format(-i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])

data_proton_UU_dn_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_UU_dn_dn_tsep_{}_sink_mom_px0_py0_pz0'.format(i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])
data_proton_UU_up_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_UU_up_up_tsep_{}_sink_mom_px0_py0_pz0'.format(i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])
data_np_proton_UU_dn_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_np_UU_dn_dn_tsep_{}_sink_mom_px0_py0_pz0'.format(-i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])
data_np_proton_UU_up_3pt = np.array([file['gf1p0_w3p0_n30_M51p2_L58_a1p5']['formfac']['ml0p0126']['proton_np_UU_up_up_tsep_{}_sink_mom_px0_py0_pz0'.format(-i)]['A3']['px0_py0_pz0']['local_curr'] for i in range(3, 13)])


                                                             
```

```python
data = gv.BufferDict()
data['2pt'] = h5data_methods.get_2pt("a12m310_a_avg.h5")
data['2pt_SS'] = data['2pt'][:,:,0]
data['2pt_PS'] = data['2pt'][:,:,1]
data['3pt'] = h5data_methods.get_3pt("a12m310_a_avg.h5", "A3", 3, 13)
```

```python
# Average the data 
# Set up the data set, average over the spin up, spin down, negative and positive parity proton, over all 1054 points
protons_2pt = (data_proton_up_2pt + data_proton_dn_2pt + data_np_proton_up_2pt + data_np_proton_dn_2pt)/4.
proton_timeavg_2pt = gv.dataset.avg_data(protons_2pt)
print("2-pt time-averaging done.")

protons_3pt = ((data_proton_UU_up_3pt - data_proton_DD_up_3pt) - (data_proton_UU_dn_3pt - data_proton_DD_dn_3pt) - (data_np_proton_UU_up_3pt - data_np_proton_DD_up_3pt) + (data_np_proton_UU_dn_3pt - data_np_proton_DD_dn_3pt))/4. 
print("3-pt time-averaging: first stage done.")
proton_timeavg_3pt = np.array([gv.dataset.avg_data(protons_3pt[i].imag) for i in range(len(range(3, 13)))])
print("3-pt time-averaging: second stage done.")

#protons_UU_3pt = (-data_proton_UU_dn_3pt + data_proton_UU_up_3pt - data_np_proton_UU_dn_3pt + data_np_proton_UU_up_3pt)/4.
#print("3-pt UU time-averaging: first stage done.")
#proton_UU_timeavg_3pt = np.array([gv.dataset.avg_data(protons_UU_3pt[i]) for i in range(len(range(3, 13)))])
#print("3-pt UU time-averaging: second stage done.")
np.shape(proton_timeavg_3pt)
```

```python
# Create the Csum
Csum = np.array([])
for i in range(3, 13):
    sumset_sdev = np.array([proton_timeavg_3pt[i-3][j].sdev for j in range(1, i)])
    sumset_sdev2 = sumset_sdev * sumset_sdev
    csum_mean = np.sum(np.array([proton_timeavg_3pt[i-3][j].mean for j in range(1, i)]))
    csum_sdev = np.sqrt(np.sum(sumset_sdev2))
    Csum = np.append(Csum, gv.gvar(csum_mean, csum_sdev))
    print("i = ", i, " range: ", proton_timeavg_3pt[i-3][1:i])
    print("element: ", Csum[i-3])
Csum
```

```python
plt.clf()
fig = plt.figure(figsize = (10,5))

ax = plt.axes()
#ax.semilogy()

ax.errorbar([i for i in range(3, 13)], [Csum[i].mean for i in range(len(range(3, 13)))], yerr=[Csum[i].sdev for i in range(len(range(3, 13)))], fmt='.')
plt.xlabel("tsep")
plt.ylabel("C^{sum}(tsep)")
plt.title("Averaged sum 3-point Correlation functions over time")
plt.legend()
plt.show()

```
