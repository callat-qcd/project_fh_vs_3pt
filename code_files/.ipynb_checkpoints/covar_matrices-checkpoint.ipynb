```python
# Now, find the correlation value drift in the old vs. new methods
percent_diffs = np.zeros([len(old_covmat3pt), len(old_covmat3pt[i])])

for i in range(len(new_covmat3pt)):
    for j in range(len(new_covmat3pt[i])):
        if (np.abs(new_covmat3pt[i][j]) < 1e-10) and (np.abs(old_covmat3pt[i][j]) < 1e-10):
             percent_diffs[i][j] = 0
        else:    
            percent_diffs[i][j] = (np.abs(old_covmat3pt[i][j]) - np.abs(new_covmat3pt[i][j]) )/((np.abs(old_covmat3pt[i][j]) + np.abs(new_covmat3pt[i][j]) )/2.)
            
```
